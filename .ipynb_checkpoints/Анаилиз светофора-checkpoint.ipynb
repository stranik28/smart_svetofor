```python
from scipy import stats
```

```python
df_just_numb['Автомобилей'] = pd.to_numeric(df_just_numb['Автомобилей'], errors='coerce')
df1_just_numb['Автомобилей'] = pd.to_numeric(df1_just_numb['Автомобилей'], errors='coerce')


group_t1 = df_just_numb['Автомобилей']
group_t2 = df1_just_numb['Автомобилей']


t_stat, p_val = stats.ttest_ind(group_t1, group_t2, equal_var=False)

# Вывод результатов
print("t-статистика:", t_stat)
print("p-значение:", p_val)
```

```python
# t-статистика - это значение, которое показывает, насколько значимо различаются средние значения двух выборок. Чем больше это значение, тем более значимы различия между выборками. В данном случае, значение t-статистики равно 13.102761642216686, что довольно большое, что говорит о том, что различия между выборками действительно существенны.

# p-значение (p-value) - это вероятность того, что различия между выборками случайны, то есть не связаны с какими-либо факторами. Если p-значение меньше заданного уровня значимости (обычно это 0.05), то мы можем считать, что различия между выборками статистически значимы. В данном случае, p-значение равно 6.63597516686988e-39 (очень близко к нулю), что говорит о том, что различия между выборками статистически значимы.
```

```python
from scipy.stats import f_oneway


f_stat, p_val = f_oneway(group_t1, group_t2)

print('F-статистика:', f_stat)
print('p-значение:', p_val)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из файла в DataFrame
df_a = df1_just_numb

# Преобразование колонки 'Время' в формат времени
df_a['Время'] = pd.to_datetime(df_a['Время'])

# Установка 'Время' как индекс
df_a.set_index('Время', inplace=True)
df_a = df_a.resample('5T').mean()
df_a['Автомобилей'] = df_a['Автомобилей'].dropna()

# Визуализация временного ряда
df_a.plot(figsize=(10, 5))

# Разложение временного ряда на тренд, сезонность и остаток
from statsmodels.tsa.seasonal import seasonal_decompose
df_a.fillna(method='ffill', inplace=True)
decomp = seasonal_decompose(df_a['Автомобилей'], model='additive', period=288)

# Визуализация разложенных компонентов временного ряда
fig, axs = plt.subplots(4, 1, figsize=(10, 10), sharex=True)
decomp.observed.plot(ax=axs[0], title='Observations')
decomp.trend.plot(ax=axs[1], title='Trend')
decomp.seasonal.plot(ax=axs[2], title='Seasonality')
decomp.resid.plot(ax=axs[3], title='Residuals')
plt.tight_layout()

# Проверка стационарности временного ряда
from statsmodels.tsa.stattools import adfuller
result = adfuller(df_a['Автомобилей'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
    
# Применение метода скользящего среднего для сглаживания временного ряда
rolling_mean = df_a['Автомобилей'].rolling(window=10).mean()
plt.plot(df_a['Автомобилей'], label='Original')
plt.plot(rolling_mean, label='Rolling Mean')
plt.legend(loc='upper left')
plt.show()

# Вычитание скользящего среднего из исходного временного ряда
df_minus_rolling_mean = df_a['Автомобилей'] - rolling_mean
df_minus_rolling_mean.dropna(inplace=True)

# Проверка стационарности вычитанного временного ряда
result = adfuller(df_minus_rolling_mean)
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
# ADF-тест (Augmented Dickey-Fuller test) является статистическим тестом, который используется для проверки стационарности временного ряда. Он позволяет проверить наличие единичных корней в ряду (корни, которые равны 1), что свидетельствует о нестационарности ряда.

# В данном случае, ADF-статистика равна -8.295958717610791, а p-value равно 4.1396240517979294e-13, что говорит о том, что мы можем отвергнуть нулевую гипотезу о нестационарности ряда. Это означает, что временной ряд стационарен с определенным уровнем значимости (обычно 5% или 1%), так как ADF-статистика находится далеко от критических значений (-3.432, -2.862, -2.567) на уровне значимости 1%, 5% и 10% соответственно.
```

```python
# Результаты ADF-теста указывают на то, что временной ряд является стационарным с высокой вероятностью. Это может говорить о том, что система в целом работает стабильно и предсказуемо, без явных изменений в характеристиках со временем. Однако, эта оценка эффективности системы может быть неполной, так как она не учитывает другие факторы, которые могут влиять на работу системы, например, изменения внешней среды, изменение потребностей пользователей или технические проблемы.
```
